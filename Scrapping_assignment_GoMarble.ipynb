# 1. Sample site 1  Recovery cream

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://2717recovery.com/products/recovery-cream")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
### SCRAPED TEXT FROM WEBSITE:
{page_data}

### INSTRUCTION

Extract all reviews from the website in the following JSON format:

{{
  "reviews_count": <total_reviews>,
  "reviews": [
    {{
      "title": "<review_title>",
      "body": "<review_body_text>",
      "rating": <rating_value>,
      "reviewer": "<reviewer_name>"
    }},
    ...
  ]
}}

### VALID JSON (NO PREAMBLE):
"""
)

chain_extract = prompt_extract | llm

# Assuming `page_data` contains the scraped text from the website
res = chain_extract.invoke(input={"page_data": page_data})
print(res.content)
```
